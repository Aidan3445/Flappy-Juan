```python
import sys
import pygame as pg
import numpy as np
import tensorflow as tf


class XOR:
    def __init__(self):
        self.w = 600
        self.h = 600
        self.res = 30
        self.cols = int(self.w / self.res)
        self.rows = int(self.h / self.res)
        self.trainInputs = tf.constant([
            [1.0, 0.0],
            [0.0, 1.0],
            [1.0, 1.0],
            [0.0, 0.0]])
        self.trainTargets = tf.constant([
            [1.0],
            [1.0],
            [0.0],
            [0.0]])
        self.testData = self.makeTestData()
        self.model = self.makeModel()
        
    def makeTestData(self):
        testArray = []
        for i in range(0, self.cols, 1):
            for j in range(0, self.rows, 1):
                x1 = i / self.cols
                x2 = j / self.rows
                testArray.append([x1, x2])
        return testArray
        
    def makeModel(self):
        model = tf.keras.Sequential()
        model.add(tf.keras.layers.Dense(10, input_shape=(2,), activation=tf.keras.activations.relu,
                                        kernel_initializer=tf.random_normal_initializer))
        model.add(tf.keras.layers.Dense(1, activation=tf.keras.activations.sigmoid,
                                        kernel_initializer=tf.random_normal_initializer))
        model.compile(tf.keras.optimizers.Adam(learning_rate=0.001), loss='mse')
        return model
    
    def train(self):
        return self.model.fit(self.trainInputs, self.trainTargets, epochs=25, batch_size=4, verbose=0, shuffle=True)

        
def run(world: XOR):
    window = pg.display.set_mode((world.w, world.h))
    c = 0
    while True:
        c += 1
        for event in pg.event.get():
            if event.type == pg.QUIT:
                print(xor.model(xor.trainInputs, training=False))
                pg.quit()
                sys.exit()
        outputArray = world.model(np.array(world.testData), training=False).numpy()
        count = 0
        for i in range(0, world.cols, 1):
            for j in range(0, world.rows, 1):
                y = int(outputArray[count][0] * 255)
                pg.draw.rect(window, (y, y, y), pg.Rect(i * world.res, j * world.res, world.res, world.res))
                count += 1
        font = pg.font.SysFont(None, 24)
        trainCount = font.render(str(c), True, (255, 0, 0))
        center = trainCount.get_rect(center=(world.w / 2, world.h / 2))
        window.blit(trainCount, center)
        pg.display.update()
        world.train()
        
        
tf.keras.backend.clear_session()
xor = XOR()
pg.init()
pg.font.init()
run(xor)
```
